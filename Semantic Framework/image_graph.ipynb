```python
import matplotlib.pyplot as plt

# Data extracted from the user input:
image_sizes = [146810, 138521, 138521, 147162, 138521, 138521, 146810, 146648]
payload_sizes = [77572, 13, 13, 13, 13, 13, 13, 13]
trials = list(range(1, 9))

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(trials, image_sizes, marker='o', label='Image-size (bytes)')
plt.plot(trials, payload_sizes, marker='s', label='Payload-size (bytes)')
plt.xlabel('Iterations')
plt.ylabel('Size (bytes)')
plt.title('Image vs Payload Size across Iterations')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
